```python
import numpy as np
from IPython.display import Audio
import matplotlib.pyplot as plt

SR = 48000
QN = 60/125/2

def gen_wav(freq_array):
    length = freq_array.size / SR
    return np.sin(np.multiply(freq_array, np.linspace(0, np.pi * 2 * length, freq_array.size)))

def note_to_freq(note):
    order = [["A"], ['A#', 'Bb'], ["B"], ["C"], ["C#", "Db"], ["D"], ["D#", "Eb"], ["E"], ["F"], ["F#", "Gb"], ["G"], ["G#", "Ab"]]
    
    parts = note.split(" ")
    name = parts[0]
    num = int(parts[1])
    
    for i, v in enumerate(order):
        if name in v:
            base_freq = 440 * (2**(1/12))**i
            break
    base_freq *= 2**(num - 4)
    return base_freq

def seq_to_arr(seq):
    ret = np.array([])
    freqs = [note_to_freq(i) for i in seq.split(",")]
    for i, v in enumerate(freqs):
        ret = np.append(ret, np.full((int(SR * QN)), v))
        if i < len(freqs) - 1:
            tlen = int(SR * QN * 4)
            # arr = v + (freqs[i + 1] - v) * (1 / (np.exp(-np.linspace(-8,8, tlen)) + 1))
            # ret = np.append(ret, arr)
            slope = (freqs[i + 1] - v) / tlen
            ret = np.append(ret, np.linspace(v, slope * freqs[i + 1], tlen))
            # if i == 0:
            #    np.savetxt("freqs.csv", arr, delimiter=",")
    # plt.plot(ret)
    return gen_wav(ret)
    
notes = seq_to_arr("D 4,D 4,D 4,B 4,D 4,D 4,G 4,G 4,F# 4,F# 4,E 4,E 4,D 4") # seq_to_arr("D 4,C 4,B 4,A 4,G 3,F# 3,E 3,D 3,C 3,D 3,E 3,F# 3,G 3")
terror = gen_wav(internal := np.append(440 / ( 1 + np.exp(-np.linspace(-8, 8, SR * 10))), np.full((int(SR * 5)), 440)))
plt.plot(internal)
audio = Audio(terror, rate=SR)

with open('test.wav', 'wb') as f:
    f.write(audio.data)
audio
```
